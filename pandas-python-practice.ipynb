```python
# What is a Series?
# A Pandas Series is like a column in a table.
# It is a one-dimensional array holding data of any type.

s=['waje',4,5]
df=pd.Series(s)
print(s)
# This label can be used to access a specified value.starts with o
print(s[1])
print(df[1])
#for series accesed in both the ways
```

```python
# Create Labels
# With the index argument, you can name your own labels.
w=[3,5,6,7,8]
df=pd.Series(w,index=['a','b','c','d','e'])
print(df)
print(df['a'])
print(w[0])
```

```python
# Key/Value Objects as Series
# You can also use a key/value object, like a dictionary, when creating a Series.
name={
    "first":"wajeeha",
    "last":"shafique",
    "age":2
}
df=pd.Series(name)
print(df)
```

```python
# To select only some of the items in the dictionary,
# use the index argument and specify only the items you want to include in the Series.
name={
    "first":"wajeeha",
    "last":"shafique",
    "age":2
}
df=pd.Series(name,index=["first","last"])
print(df)
```

```python
# DataFrames
# Data sets in Pandas are usually multi-dimensional tables, called DataFrames.
# Series is like a column, a DataFrame is the whole table.

data={
    "day":['d1','d2','d3'],
    "cal":[435,678,965]
}
df=pd.DataFrame(data)
print(df)
```

```python
# Locate Row
# As you can see from the result above, the DataFrame is like a table with rows and columns.
# Pandas use the loc attribute to return one or more specified row(s)
print(df.loc[0]) #refer to the row index:
print(df.loc[0][1])
# Return row 0 and 1
print(df.loc[[0,1]])
```

```python
# Named Indexes
# With the index argument, you can name your own indexes.
df=pd.DataFrame(data,index=["a","b","c"])
print(df)

# Locate Named Indexes
# Use the named index in the loc attribute to return the specified row(s).
print(df.loc["a"])
```

```python
# Pandas - Cleaning Empty Cells
# Empty cells can potentially give you a wrong result when you analyze data.

# methods:
# 1.remove rows that contain empty cells.....for very big data removing a few rows will not have a big impact on the result.
newdf=df.dropna()
print(newdf.to_string())

# By default, the dropna() method returns a new DataFrame, and will not change the original.

# If you want to change the original DataFrame, use the inplace = True argument

# df.dropna(inplace=true)
# print(df)

# Now, the dropna(inplace = True) will NOT return a new DataFrame,
# but it will remove all rows containing NULL values from the original DataFrame.
```

```python
newdf.info()
#now there are no null  values...in calories 
```

```python
# 2.Replace Empty Values...insert a new value instead... and in this way not have to delete entire row
# The fillna() method ...allows us to replace empty cells with a value

# syntax......dataframe.fillna(value, method, axis, inplace, limit, downcast)

# Replace NULL values with the number 130
df.fillna(130, inplace = True)
df.info()
```

```python
# Mean = the average value (the sum of all values divided by number of values).
# Median = the value in the middle, after you have sorted all values ascending.
# Mode = the value that appears most frequently.
```

```python
# Replace Only For Specified Columns
df = pd.read_csv('data.csv')

# Replace Using Mean, Median, or Mode
# there are functions in pandas...min,max,mode,median,mean...all are accessed in the same way
m=df["Calories"].mean()
print(m,"\n")

# Replace NULL values in the "Calories" columns with mean value on column Calories
df["Calories"].fillna(m,inplace=True)
df.info() #all null value are replaced with mean valus
```

```python
# 1.Replacing Values
# Set "Duration" = 45 in row 7

# example
df.loc[7,"Duration"]=60
df.head(8)
```

```python
# # For small data sets you might be able to replace the wrong data one by one, but not for big data sets.

# Loop through all values in the "Duration" column.
# If the value is higher than 55, set it to 55

for x in df.index:
    if df.loc[x, "Duration"] > 55:
        df.loc[x, "Duration"] = 55

        
print(df)
```

```python
# 2.Removing Rows
# Delete rows where "Duration" is higher than 55

for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.drop(x, inplace = True)
        
print(df)        
```

```python
# The duplicated() method returns a Boolean values for each row
# Returns True for every row that is a duplicate, othwerwise False
df=pd.read_csv("data.csv")
duplicated = df.duplicated()
sum(~df.duplicated())
```

```python
# Removing Duplicates
# use the drop_duplicates() method.
df.drop_duplicates(inplace=True)

#  The (inplace = True) will make sure that the method does NOT return a new DataFrame, but it will remove all duplicates from the original DataFrame.
sum(~df.duplicated())
```

```python
# Pandas - Data Correlations
# The corr() method: 
# calculates the relationship between each column in your data set.
df.corr()
# The corr() method ignores "not numeric" columns.

# Result of the corr() method:
# table with a lot of numbers that represents how well the relationship is between two columns.
# The number varies from -1 to 1.

-> 1 : 1 to 1 relationship (a perfect correlation)
-> 0.9 : good relationship, and if you increase one value, the other will probably increase as well.
-> -0.9 : would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.
-> 0.2 : NOT a good relationship, meaning that if one value goes up does not mean that the other will.
```
